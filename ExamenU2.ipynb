```python
import pandas as pd
```

```python
juegos = pd.read_csv("VGSales.csv")
```

```python
juegos
```

```python
accion_aventura = juegos[(juegos['Genre'] == 'Action') | (juegos['Genre'] == 'Adventure')]
```

```python
conteo = accion_aventura.groupby(['Plataforma', 'Genre']).size().reset_index(name='Count')
```

```python
print(conteo)
```

```python
juegossuperiores = juegos[juegos['Year'] >= 2010]
```

```python
conteo = juegossuperiores.groupby(['Genre', 'Year']).size().reset_index(name='Count')
```

```python
print(conteo)
```

```python
ventasgenero = juegos.groupby('Genre')['Global_Sales'].sum().reset_index()

```

```python
print(ventasgenero)
```
